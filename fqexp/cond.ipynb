```python
import matplotlib.pyplot as plt
import math
import numpy as np

kH2O = 2.57e-6 # S/cm
concHAc = [0.0001, 0.001, 0.01, 0.05, 0.1] # mol/L
kHAc = [418.2e-6 - kH2O, 308.7e-6 - kH2O, 138.2e-6 - kH2O, 41.36e-6 - kH2O, 11.91e-6 - kH2O] # S/cm

concNaCl = [0.0001, 0.001, 0.01, 0.05, 0.1] # mol/L
kNaCl = [8.46e-3 - kH2O, 4.46e-3 - kH2O, 100e-6 - kH2O, 122.6e-6 - kH2O, 63.77e-6 - kH2O] # S/cm
# kNaCl.reverse()
print(kNaCl)
CNaCl = [math.sqrt(i) for i in concNaCl]
LmNaCl = [i for i in kNaCl]
concNaCl.reverse()
print(LmNaCl)
for i in range(len(kNaCl)):
    LmNaCl[i] = LmNaCl[i] / (concNaCl[i] * 0.001)
print(CNaCl)
print(LmNaCl)
lin = np.linalg.lstsq(np.vstack([CNaCl, np.ones(len(CNaCl))]).T, LmNaCl, rcond=None)[0]
print(lin)
plt.plot(CNaCl, LmNaCl)
plt.xlabel("Concentração^1/2")
plt.ylabel("Cond molar  S cm^2 / mol")
plt.show
```
