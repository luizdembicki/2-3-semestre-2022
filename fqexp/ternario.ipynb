```python
import ternary

# Densidades, g/cm^3 = g/mL 
dens = {
    "AcOEt" : 0.897, 
    "EtOH" : 0.789,
    "H2O" : 0.997
}
# Massa molar g / mol
MM = { 
    "AcOEt" : 88.11,
    "EtOH" : 46.07,
    "H2O" : 18.01528 
}
# inicializar os tubos
tubo = [[0 for i in range(3)] for x in range(18)]
for i in range(18):
    # inserir valores e mudar para mols
    for x in range(0, 3):
        if x == 0:
            tubo[i][x] = (dens["H2O"] * float(input(f'Tubo {i + 1} H2O:').replace(',', '.'))) / MM["H2O"]
        elif x == 1:
            tubo[i][x] = (dens["AcOEt"] * float(input(f'Tubo {i + 1} AcOEt:').replace(',', '.'))) / MM["AcOEt"]
        else:
            tubo[i][x] = (dens["EtOH"] * float(input(f'Tubo {i + 1} EtOH:').replace(',', '.'))) / MM["EtOH"]
    # mudar para fração molar
    molar = tubo[i][0] + tubo[i][1] + tubo[i][2] 
    for x in range(3):
        tubo[i][x] = tubo[i][x] / molar
        print("Tubo", i, x, ":",tubo[i][x])

    
```

```python
# escala
scale = 1
figure, tax = ternary.figure(scale=scale)

# bordas e grid
tax.boundary(linewidth=2.0)
tax.gridlines(color="black", multiple=0.05)


# Eixos e título
fontsize = 20
tax.set_title("Diagrama Ternario H2O-AcOEt-EtOH", fontsize=fontsize)
tax.bottom_axis_label("H2O", fontsize=fontsize, offset=0.05) # x
tax.right_axis_label("AcOEt", fontsize=fontsize, offset=0.12) # y
tax.left_axis_label("EtOH", fontsize=fontsize, offset=0.12) # z

# Contadores laterais
tax.ticks(axis='lbr', multiple=0.1, linewidth=1, offset=0.017, tick_formats="%.1f")

# Remover o default
tax.clear_matplotlib_ticks()
# carregar dado em tuple (x,y,z)
points = tuple(tubo)
# Plotar
tax.plot(points, linewidth=2.0, label="Curva Limite")
tax.scatter(points, color='green', label="Pontos")
tax.legend()
tax.get_axes().axis('off')
tax.show()

```
